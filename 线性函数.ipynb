# 线性函数
$f(x) = a^{T}x$ 其中 $a是n维向量$

因此可以推出
$f(\alpha x + \beta y) = \alpha f(x) + \beta f(y)$

```julia
a = [-2, 0, 1, -3]
```

```julia
f(x) = a' * x
```

```julia
x = [2, 2, -1, 1]; y = [0, 1, -1, 0];
```

```julia
α = 1.5; β = -3.7;
```

```julia
lhs = f(α * x + β * y)
```

```julia
rhs = α*f(x) + β*f(y)
```

```julia
avg(x) = (ones(length(x)) / length(x))' * x
```

```julia
x = [1, 3, 2, -1]
avg(x)
```

## 泰勒展开
一阶近似 $f:R^{n}->R$
    $$\hat{f}(x) = f(z) + \nabla f(z)^{T}(x-z)$$

```julia
f(x) = x[1] + exp(x[2]-x[1])
```

```julia
grad_f(z) = [1 - exp(x[2] - x[1]), exp(x[2] - x[1])] # x[1]的导， x[2]的导
```

```julia
z = [1, 2]
```

```julia
f_hat(x) = f(z) + grad_f(z)' * (x-z)
```

```julia
f(z), f_hat(z)
```

```julia
z = [1, 2]; z1 = [0.9, 1.9]
```

```julia
f(z), f_hat(z1)
```

## 回归模型
 $f(x)=x^{T}\beta+v$      
 其中$\beta$是$n$维向量$v$是标量, 它们是模型中的参数，这个模型被用来通过一组$x$来近似$y$值

```julia
β = [148.73, -18.85]; v = 54.40;
```

```julia
y_hat(x) = x' * β + v
```

```julia
x = [0.846, 1]; y = 115;
```

```julia
y_hat(x), y
```

```julia
x = [1.324, 2]; y = 234.50;
```

```julia
y_hat(x), y
```
